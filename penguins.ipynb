##  2.B Handling Missing Values

```python
# Impute missing values with mean
data['culmen_length_mm'].fillna(data['culmen_length_mm'].mean(), inplace=True)
data['culmen_depth_mm'].fillna(data['culmen_depth_mm'].mean(), inplace=True)
data['flipper_length_mm'].fillna(data['flipper_length_mm'].mean(), inplace=True)
data['body_mass_g'].fillna(data['body_mass_g'].mean(), inplace=True)
```

### 2.B.I.1. Imputation for Numerical Columns (culmen_length_mm, culmen_depth_mm, flipper_length_mm, body_mass_g):

**Strategy:**
- Since the missing values are relatively small in number (2 for each column), Let's consider imputing them with the mean of the respective columns.

**Explanation:**
- The missing values in numerical columns ('culmen_length_mm', 'culmen_depth_mm', 'flipper_length_mm', 'body_mass_g') are addressed through imputation.
- Imputing with the mean a common approach when dealing with small amounts of missing data.
- It ensures that the imputed values align with the central tendency of the existing data distribution.


```python
# Impute missing values in 'sex' with the mode
data['sex'].fillna(data['sex'].mode()[0], inplace=True)
```

### 2.B.I.2. Handling Missing Values for Categorical Column (sex):

**Strategy:**
- Since 'sex' is a categorical variable, Let us choose to fill the missing values with the mode (most frequent category) as it maintains the categorical nature of the variable.

**Explanation:**
- The missing values in the 'sex' column, representing the gender of the penguins, are handled considering its categorical nature.
- Filling missing values with the mode ensures that the imputed values represent the most frequently occurring gender.
- Alternatively, analyzing the distribution of 'sex' can provide insights into imputing missing values based on the existing proportions of males and females.
- This strategy is appropriate for maintaining the categorical integrity of the 'sex' variable, which would be essential for downstream analyses.

```python
# Convert 'sex' column to categorical data type
data['sex'] = data['sex'].astype('category')
```
